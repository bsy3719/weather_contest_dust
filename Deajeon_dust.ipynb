```python
import pandas as pd
import numpy as np
```

```python
dust2018 = pd.read_csv("deajeon_dust/2018_dust.csv", sep= ",", dtype='unicode', encoding='euc-kr')
dust2017 = pd.read_csv("deajeon_dust/2017_dust.csv", sep= ",", dtype='unicode', encoding='euc-kr')
```

```python
dust = pd.concat([dust2018, dust2017])
```

```python
dust.isnull().sum()
```

```python
dust.count()
```

```python
drop_dust = dust.drop_duplicates(['Datetime'])
```

```python
drop_dust.count()
```

```python
weather2018 = pd.read_csv("deajeon_weather/weather_2018.csv", sep= ",", dtype='unicode', encoding='euc-kr')
weather2017 = pd.read_csv("deajeon_weather/weather_2017.csv", sep= ",", dtype='unicode', encoding='euc-kr')
```

```python
weather = pd.concat([weather2018, weather2017])
```

```python
weather.head()
```

```python
weather.count()
```

```python
weather.isnull().sum()
```

```python
drop_dust['Datetime'] = drop_dust['Datetime'].astype(str)
```

```python
drop_dust['year'] = drop_dust['Datetime'].str[0:4]
drop_dust['month'] = drop_dust['Datetime'].str[4:6]
drop_dust['day'] = drop_dust['Datetime'].str[6:8]
drop_dust['time'] = drop_dust['Datetime'].str[8:10]
drop_dust.drop(['Datetime'], axis=1, inplace=True)
```

```python
weather.head()
```

```python
weather['일시'] = weather['일시'].astype(str)
```

```python
weather['year'] = weather['일시'].str[0:4]
weather['month'] = weather['일시'].str[5:7]
weather['day'] = weather['일시'].str[8:10]
weather['time'] = weather['일시'].str[11:13]
weather.drop(['일시'], axis=1, inplace=True)
```

```python
weather.head()
```

```python
weather.loc[weather['time'].str.contains('1:'), 'time'] = '01'
weather.loc[weather['time'].str.contains('2:'), 'time'] = '02'
weather.loc[weather['time'].str.contains('3:'), 'time'] = '03'
weather.loc[weather['time'].str.contains('4:'), 'time'] = '04'
weather.loc[weather['time'].str.contains('5:'), 'time'] = '05'
weather.loc[weather['time'].str.contains('6:'), 'time'] = '06'
weather.loc[weather['time'].str.contains('7:'), 'time'] = '07'
weather.loc[weather['time'].str.contains('8:'), 'time'] = '08'
weather.loc[weather['time'].str.contains('9:'), 'time'] = '09'
```

```python
drop_dust['year'] = drop_dust['year'].astype(str)
drop_dust['month'] = drop_dust['month'].astype(str)
drop_dust['day'] = drop_dust['day'].astype(str)
drop_dust['time'] = drop_dust['time'].astype(str)

weather['year'] = weather['year'].astype(str)
weather['month'] = weather['month'].astype(str)
weather['day'] = weather['day'].astype(str)
weather['time'] = weather['time'].astype(str)
```

```python
merged_dust = drop_dust.merge(weather, on = ['year', 'month', 'day', 'time'])
```

```python
merged_dust.head()
```

```python
merged_dust.rename(columns={'기온(°C)': "Temperature", '풍향(deg)' : 'Wind_direction', '풍속(m/s)' : 'Wind_speed', '강수량(mm)' : 'Precipitation', '습도(%)' : 'Humidity'}, inplace =True)
```
